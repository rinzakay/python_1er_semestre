**Conociendo la BIBLIOTECA Matplotlib**

```python
import matplotlib.pyplot as plt
```

creando un grafico simple

```python
plt.plot(vector)
plt.title("Grafico de vectores")
plt.xlabel("Indice (posicion)") #Para agregar el eje x al grafico
plt.ylabel("Valor")#Para

plt.grid(True)
```

**Crear dos vectores en un mismo grafico**

```python
vector_1 = np.random.randint(1,11,12)
vector_2 = np.random.randint(2,20,12)
```

```python
plt.plot(vector_1, marker = 'o', label = vector_1)

plt.title("Comparando dos vectores")
plt.xlabel("Indice (posicion)")
plt.ylabel("Valor")

plt.grid(True)
plt.legend()

plt.show()
```

**Creando un Gr√°fico de barra**
