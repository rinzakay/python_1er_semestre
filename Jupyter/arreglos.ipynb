```python
import numpy as np
```

Crear una lista de 10 elementos

```python
lista = list(range(1,11))
lista
```

transformando la lista a un array/arreglo

```python
arreglo = np.array(lista)
arreglo
```

Para calcular el promedio

```python
promedio_int = np.mean(lista)
promedio_int
```
